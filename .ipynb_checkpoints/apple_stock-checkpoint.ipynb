```python
#Target is whether tomorrow's close is higher than today's close
data['tomorrow'] = data['Close'].shift(-1)

data['target'] = (data['tomorrow'] > data['Close']).astype(int)
```

MODEL DEVELOPMENT

```python
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import precision_score
```

```python
predictors = ['Open', 'High', 'Low', 'Close', 'Volume']

train = data.iloc[:-200]
test = data.iloc[-200:]
```

```python
print(train.shape)
print(test.shape)
```

```python
#Test different estimators to obtain the best.

model_params = {
    "knn" : {
        "model" : KNeighborsClassifier(n_jobs=2),
        "params" : {
            "n_neighbors" : [5,10, 15],
            "leaf_size" : [15, 20, 30,40],
            "weights" : ['uniform', 'distance']
        }
    },

    "hst" : {
        "model" : HistGradientBoostingClassifier(random_state=42),
        "params" : {
            "max_iter" : [75, 100, 150, 200],
            "max_depth" : [5, 10,15,20],
            "max_features" : [0.2, 0.5, 0.7, 1.0],
            "min_samples_leaf": [10,15,20,30]
        }
    },

    "rfc" :{
        "model" : RandomForestClassifier(random_state=42, n_jobs=2, class_weight='balanced'),
        "params" : {
            "n_estimators" : [50, 100, 200, 300],
            "max_depth": [5,10,15,20],
            "max_leaf_nodes": [2,5,7],
            "min_samples_leaf": [2,4,7,10]
        }
    }
}
```

```python
tscv = TimeSeriesSplit(n_splits=5)

scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=tscv, return_train_score=True)
    clf.fit(train[predictors], train['target'])
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
scores
```

```python
model = HistGradientBoostingClassifier(max_depth=5, min_samples_leaf=10, random_state=42)

model.fit(train[predictors], train['target'])
```

```python
y_preds = model.predict(test[predictors])
y_preds = pd.Series(y_preds, index=test.index)

print(f'model precision: ', precision_score(test['target'], y_preds, zero_division=1))
```

BACKTESTING

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name='predicted')
    combined = pd.concat([test['target'], preds], axis =1)
    return combined
    
```

```python
def backtest(data, model, predictors,start=500, step=250):  #every trading year is about 250 days
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions  = backtest(data, model, predictors)
```

```python
predictions['predicted'].value_counts()
```

```python
precision_score(predictions['target'], predictions['predicted'])
```

```python
predictions['target'].value_counts() / predictions.shape[0]

#the market went up only 47% of the days under review
```

IMPROVING THE MODEL

```python
#add more features and data
horizons = [2,5,20,60,120,240]

new_predictors = []

for horizon in horizons:
    rolling_average = data.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    data[ratio_column] = data['Close'] / rolling_average['Close']

    trend_column = f"trend_{horizon}"
    data[trend_column] = data.shift(1).rolling(horizon).sum()['target']

    new_predictors += [ratio_column, trend_column] + predictors # add the initials predictors
```

```python
data.head(10)
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
new_predictors
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['target'])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name='predicted')
    combined = pd.concat([test['target'], preds], axis =1)
    return combined
    
```

```python
new_predictions = backtest(data, model, new_predictors)
```

```python
new_predictions['predicted'].value_counts()
```

```python
precision_score(new_predictions['target'], new_predictions['predicted'])

#precision is 59% as against 47% buys of the total trading days
```
